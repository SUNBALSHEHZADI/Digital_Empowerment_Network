```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['GPA'].fillna(df['GPA'].mean(), inplace=True)
df['Python'].fillna(df['Python'].ffill(), inplace=True)
df['SQL'].fillna(df['SQL'].ffill(), inplace=True)
df['Java'].fillna(df['Java'].ffill(), inplace=True)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Major'].fillna(df['Major'].mode()[0], inplace=True)
df['Interested Domain'].fillna(df['Interested Domain'].mode()[0], inplace=True)
```

```python
df.dropna(subset=['Age', 'GPA', 'Python', 'SQL', 'Java'], inplace=True)
```
